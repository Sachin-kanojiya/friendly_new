```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(response.text)
print(soup.prettify())
```

```python
soup.find("title")
```

```python
soup.find_all("article")
```

```python
pageinfo= soup.find_all(class_="page") # it is may possible header is in container ;class ;id ect 
    
for r in pageinfo:
     print(r.get_text(strip=True))
     
```

```python
reviews = soup.find_all("p", class_="review-text")

for r in reviews:
    print(r.get_text(strip=True))
```

```python
# for bussiness phone no 
print(soup.find("span",class_="phone").text)
```

```python
elem = soup.find(class_="main")

if elem:
    print(elem.get_text(strip=True))
```

```python
# find all the link on the web page 

link=soup.find_all("a")

for links in link:
    print(links.text,links.get("herf"))
```
